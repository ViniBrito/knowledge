```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from scipy import ndimage

subplot_count = 7
path = 'mickey.jpg'
# img = cv2.imread(path) # Equivale √†s duas linhas seguintes
img = Image.open(path)
img = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

# Imagem original
plt.figure(figsize=(20, 15)) # J√° serve pra configurar as dimens√µes de todos os subplots
plt.subplot(1, subplot_count, 1)
plt.title("Original")
plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# Escala de cinza
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.subplot(1, subplot_count, 2)
plt.title("Escala de cinza")
plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
cv2.imwrite('grayscale.png', img)
# img = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 20)
# img = cv2.GaussianBlur(img, (5,5), 0)

# Degrau bin√°rio
img = cv2.threshold(img, 195, 255, cv2.THRESH_BINARY)[1]
plt.subplot(1, subplot_count, 3)
plt.title("Degrau bin√°rio")
plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))
cv2.imwrite("threshold.png", img)

# Rota√ß√£o
angle = -8.5
img = ndimage.rotate(img, angle)
plt.subplot(1, subplot_count, 4)
plt.title(f"Rotacionada em {angle}¬∞")
plt.axis('off')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_GRAY2RGB))

# Invers√£o
plt.subplot(1, subplot_count, 5)
plt.title("Invertido")
plt.axis('off')
plt.imshow(cv2.cvtColor(255-img, cv2.COLOR_GRAY2RGB))
plt.show()

# Corte vertical
plt.subplot(1, subplot_count, 6)
plt.title("Corte vertical")
plt.axis('off')
plt.imshow(cv2.cvtColor(img[:, 100:400], cv2.COLOR_GRAY2RGB))
plt.show()

# Corte horizontal
plt.subplot(1, subplot_count, 7)
plt.title("Corte horizontal")
plt.axis('off')
plt.imshow(cv2.cvtColor(img[100:400, :], cv2.COLOR_GRAY2RGB))
plt.show()
```
